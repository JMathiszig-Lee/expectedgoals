```python
import pandas as pd
```

```python
prem = pd.read_csv("prem.csv")
prem.head()
```

```python
prem.describe()
```

```python
import scipy

slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(prem["xG"], prem["Goals"])
print(r_value)
```

```python
prem['Goals_Predicted'] = prem['xG'] * slope + intercept
prem.head()
```

```python
from bokeh.io import push_notebook, show, output_notebook, export_png
from bokeh.plotting import figure, show, save, output_file
from bokeh.models import Range1d

output_notebook()
output_file('xg.html')
```

```python
plot = figure(title=f"xG vs Actual Goals", tools="hover, save", tooltips="@Team: @Game")
plot.circle('xG', 'Goals', source=prem, fill_alpha=0.3, size=10, legend_label=f"r={r_value:.4}")
plot.line('xG', 'Goals_Predicted', source=prem, color="red", legend_label=f"best fit: Goals = xG * {slope:.2} + {intercept:.2}")
plot.x_range = Range1d(0, 10)
plot.y_range = Range1d(0, 15)
plot.xaxis.axis_label = 'xG'
plot.yaxis.axis_label = 'Goals'
show(plot)
save(plot)
```

Look at just the world cup

```python
output_file('worldcup.html')
slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(prem["xG"][:51], prem["Goals"][:51])


plot = figure(title= "xG vs Actual Goals at the 2022 world cup", tools="hover, save", tooltips="@Team: @Game")
plot.circle('xG', 'Goals', source=prem[:51], color="green", fill_alpha=0.3, size=10, legend_label=f"r={r_value:.4}")
plot.x_range = Range1d(-1, 10)
plot.y_range = Range1d(-1, 10)
plot.xaxis.axis_label = 'xG'
plot.yaxis.axis_label = 'Goals'
plot.yaxis.fixed_location = 0
plot.xaxis.fixed_location = 0
show(plot)
save(plot)
```
