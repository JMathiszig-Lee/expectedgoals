```python
import pandas as pd
```

```python
prem = pd.read_csv("prem.csv")
prem.head()
```

```python
prem.describe()
```

```python
import scipy

slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(prem["xG"], prem["Goals"])
print(r_value)
```

```python
prem['Goals_Predicted'] = prem['xG'] * slope + intercept
prem.head()
```

```python
from bokeh.io import push_notebook, show, output_notebook, export_png
from bokeh.plotting import figure, show
from bokeh.models import Range1d

output_notebook()

```

```python
plot = figure(title= "xG vs Actual Goals", tools="hover", tooltips="@Team: @Game")
plot.circle('xG', 'Goals', source=prem, fill_alpha=0.5, size=10, legend=f"r={r_value:.4}")
plot.line('xG', 'Goals_Predicted', source=prem, color="red", legend=f"slope of best fit")
plot.x_range = Range1d(0, 10)
plot.y_range = Range1d(0, 15)
plot.xaxis.axis_label = 'xG'
plot.yaxis.axis_label = 'Goals'
show(plot)
```

```python
from bokeh.plotting import save
save(plot)
```

```python
export_png(plot, filename='xgtruth.png')
```
