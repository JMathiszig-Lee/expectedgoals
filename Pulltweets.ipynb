```python
import re

reg_float="[0-9]+\.[0-9]+"
reg_score="[0-9]+-[0-9]+"
```

```python
for i in xg_tweets[0]:
    i=str(i)
    print(i)
    match = re.search(reg_float, i)
    print(match)
    print(match.start())
    print(i[:match.start()-1])
    print("-"*20)
```

```python
for i in xg_tweets[0]:
    i=str(i)
    print(i)
    if re.match("^HT:", i):
        pass
    else:
        match = re.findall(reg_float, i)
        if len(match) > 1:
            first_match = re.search(match[0],i)
            second_match = re.search(match[1], i)
            score_match = re.search(reg_score, i)
            if first_match and second_match and score_match:
                print(i[:first_match.start()-1])
                print(match[0])
                print(score_match.group()[:1])
                print(score_match.group()[2:])
                print(match[1])
                print(i[second_match.end()+2:])
        
    print("-"*20)
```

```python
xg_tweets = []

for tweet in tweepy.Paginator(client.get_users_tweets, xg_id, exclude="retweets", max_results=100).flatten():
    xg_tweets.append(tweet)
```

```python
import csv
headers = ["Team", "xG", "Goals", "Home/Away", "Game"]
with open('prem.csv', 'w', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(headers)

    for i in xg_tweets:
        i=str(i)
        if re.match("^HT:", i):
            #pass if halftime update
            pass
        elif i.count('\n') >0:
            #pass if not 1 line format
            pass
        else:
            match = re.findall(reg_float, i)
            if len(match) > 1:
                first_match = re.search(match[0],i)
                second_match = re.search(match[1], i)
                score_match = re.search(reg_score, i)

                if first_match and second_match and score_match:
                    team0 = i[:first_match.start()-2]
                    score0 = score_match.group()[:1]
                    team1 = i[second_match.end()+2:]
                    score1 = score_match.group()[2:]
                    game = f"{team0} {score0}-{score1} {team1}"
                    writer.writerow([team0, match[0], score0, "Home", game])
                    writer.writerow([team1, match[1], score1, "Away", game])
                    print(game)
```
